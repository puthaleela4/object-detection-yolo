```python
import glob

train_lbl_dir = os.path.join(DATASET_ROOT, "train", "labels")
sample_label_file = glob.glob(train_lbl_dir + "/*.txt")[0]
print("Sample label file:", sample_label_file)
with open(sample_label_file, "r") as f:
    for line in f.readlines():
        print(line.strip())
```

```python
import os
import glob

DATASET_ROOT = "/content/drive/MyDrive"

def parse_labels(lbl_dir):
    ids = []
    for file in glob.glob(lbl_dir + "/*.txt"):
        try:
            with open(file) as f:
                for line in f:
                    parts = line.strip().split()
                    # Check if the line has at least one element before attempting to convert to int
                    if parts:
                        try:
                            ids.append(int(parts[0]))
                        except ValueError:
                            print(f"Warning: Could not convert '{parts[0]}' to an integer in file {file}")
        except Exception as e:
            print(f"Error reading file {file}: {e}")
    return ids

# Redefine train_lbl_dir
train_lbl_dir = os.path.join(DATASET_ROOT, "train", "labels")

all_ids = parse_labels(train_lbl_dir)
unique_classes = sorted(set(all_ids))

if not unique_classes:
    print("No unique class IDs found in label files. Please check your label file format.")
else:
    print("Unique class IDs:", unique_classes)
```

```python
class_names = {0: "ruins", 1: "vegetation", 2: "artifact"}
for i in unique_classes:
    print(f"Class {i}: {class_names.get(i, 'Unknown')}")
```

```python
yaml_content = f"""
train: {DATASET_ROOT}/train
val: {DATASET_ROOT}/valid
test: {DATASET_ROOT}/test

nc: {len(class_names)}
names: {list(class_names.values())}
"""

yaml_path = os.path.join(DATASET_ROOT, "data.yaml")
with open(yaml_path, "w") as f:
    f.write(yaml_content)

print("data.yaml contents:\n")
print(open(yaml_path).read())
```

```python
!pip install ultralytics
```

```python
from ultralytics import YOLO

# Load pre-trained YOLOv8 nano model
model = YOLO("yolov8n.pt")

# Start training
model.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=8,
    name="archaeology_yolov8",
    project="/content/drive/MyDrive/YOLO_Training",
    workers=2
)
```

```python
# Validate model performance
model.val()
```
